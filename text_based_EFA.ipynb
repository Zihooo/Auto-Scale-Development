```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
embeddings = model_raw.encode(statements, convert_to_numpy=True)
# ----- 3. Cosine‑similarity matrix -------------------------------------------
sim_matrix = cosine_similarity(embeddings)

# ----- 4. Collect pairs above threshold --------------------------------------
threshold = 0.80
pairs = []
n = len(statements)

for i in range(n):
    for j in range(i + 1, n):                # upper triangle, no self‑pairs
        score = sim_matrix[i, j]
        if score > threshold:
            pairs.append(
                {"idx_1": i, "idx_2": j,
                 "sent_1": statements[i],
                 "sent_2": statements[j],
                 "similarity": round(float(score), 3)}
            )

# Optional: view as a DataFrame sorted by similarity
similar_pairs_df = (
    pd.DataFrame(pairs)
      .sort_values(by="similarity", ascending=False)
      .reset_index(drop=True)
)

print(f"{len(similar_pairs_df)} pairs found (similarity > {threshold})")
print(similar_pairs_df[["idx_1", "idx_2", "similarity"]].head())
```
